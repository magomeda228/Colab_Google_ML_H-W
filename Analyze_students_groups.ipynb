```python
df = pd.read_excel('end_seminar.xlsx')
```

```python
df = pd.DataFrame(df)
```

```python
str_col = df.select_dtypes(include=['object', 'string']).columns
num_col = df.select_dtypes(include=['int64','float64']).columns
df[num_col]  = df[num_col].fillna(0)
df[str_col] = df[str_col].fillna('False')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['day']=df['timestamp'].dt.day
```

```python
sns.barplot(df.groupby(['day'])['rating'].mean())
plt.grid(True,linestyle='--',alpha=0.7)
plt.ylabel('Mean rating')
plt.title('Mean rating from day')
plt.savefig('plot.pdf', format='pdf')
```
